### 2. Measures of Dispersion  : Degree of Variation (Spread, Variability)
* Range
* Inter quartile Range
* Quartile
* Percentile
* Variance
* Standard Deviation 

```python
import numpy as np

# Sample dataset
dataset = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]

# Range : max - min (Sensitive to extreme values) , it doesn't consider the distribution, Suitable for small Dataset.
range_value = np.max(dataset) - np.min(dataset)
print(f"Range: {range_value}")
```

```python
# Interquartile Range  (IQR) : less sensitive to extreme values [IQR =Q3 - Q1 = 50% of  the Data]
Q1 = np.percentile(dataset, 25)
Q3 = np.percentile(dataset, 75)
IQR = Q3 - Q1
print(f"Interquartile Range (IQR): {IQR}")
```

```python
# Quartiles calculation using numpy.percentile:
# Q1: 25%, Q2 (Median): 50%, Q3: 75%
quartiles = np.percentile(dataset, [25, 50, 75])
print(f"Quartiles: Q1: 25% = {quartiles[0]}, Q2: 50% (Median) = {quartiles[1]}, Q3: 75% = {quartiles[2]}")
```

```python
# Percentiles : 
percentile_value = np.percentile(dataset, 90)
print(f"90th Percentile: {percentile_value}")
```

```python
# Variance : 
variance_value = np.var(dataset)
print(f"Variance: {variance_value}")
```

```python
# Standard Deviation : SQRT of Variance
std_deviation_value = np.std(dataset)
print(f"Standard Deviation: {std_deviation_value}")
```

### 3. Frequency : No of times a value of data occurred , 
* Relative Frequency
* Cumulative Frequency

```python
import numpy as np

# Sample dataset
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Frequency: Number of times each value occurs in the dataset
unique_values, counts = np.unique(data, return_counts=True)
frequency = dict(zip(unique_values, counts))
print(f"Frequency: {frequency}")
```

```python
# Relative Frequency: Proportion of times each value occurs relative to the total number of observations
total_observations = len(data)
relative_frequency = {value: count / total_observations for value, count in frequency.items()}
print(f"Relative Frequency: {relative_frequency}")
```

```python
# Cumulative Frequency: Running total of the frequencies
cumulative_frequency = np.cumsum(list(counts))
cumulative_frequency_dict = dict(zip(unique_values, cumulative_frequency))
print(f"Cumulative Frequency: {cumulative_frequency_dict}")
```

### Types of Skewed Histograms

1. **Positively Skewed (Right-skewed):**
   - In a positively skewed histogram, the right tail (the larger values) is longer than the left tail (the smaller values).
   - The majority of the data points are concentrated on the left side, and there are a few larger values on the right side.
   - The mean is typically greater than the median in a positively skewed distribution.


2. **Negatively Skewed (Left-skewed):**
   - In a negatively skewed histogram, the left tail is longer than the right tail.
   - The majority of the data points are concentrated on the right side, and there are a few smaller values on the left side.
   - The mean is typically less than the median in a negatively skewed distribution.


3. **Symmetric (No skewness):**
   - In a symmetric histogram, the distribution is balanced, and the left and right tails are of equal length.
   - The mean and median are approximately equal in a symmetric distribution.
   - There is no clear trend leaning towards larger or smaller values.



Understanding the skewness of a histogram is essential for interpreting the distribution of data. It provides insights into the concentration of values and the presence of outliers in a dataset.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Create a figure with three subplots in a single row
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Example 1: Positively Skewed (Right-skewed)
positively_skewed_data = np.random.exponential(scale=1.5, size=1000)
axes[0].hist(positively_skewed_data, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Positively Skewed (Right-skewed)')

# Example 2: Negatively Skewed (Left-skewed)
negatively_skewed_data = -np.random.exponential(scale=1.5, size=1000)
axes[1].hist(negatively_skewed_data, bins=30, color='salmon', edgecolor='black')
axes[1].set_title('Negatively Skewed (Left-skewed)')

# Example 3: Symmetric (No skewness)
symmetric_data = np.random.normal(loc=0, scale=1, size=1000)
axes[2].hist(symmetric_data, bins=30, color='lightgreen', edgecolor='black')
axes[2].set_title('Symmetric Distribution')

# Set common labels for all subplots
for ax in axes:
    ax.set_xlabel('Values')
    ax.set_ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

### Types of Histograms Based on Number of Modes

1. **Unimodal Distribution:**
   - A unimodal distribution has a single peak or mode.
   - The data is concentrated around one central value.
   - Example: Normal distribution.

2. **Bimodal Distribution:**
   - A bimodal distribution has two distinct peaks or modes.
   - The data has two different central values, creating two concentration points.
   - Example: Mixture of two normal distributions.

3. **Multimodal Distribution:**
   - A multimodal distribution has more than two peaks or modes.
   - The data has multiple concentration points, indicating different groups or patterns.
   - Example: Mixture of several normal distributions or complex real-world datasets.

4. **Uniform Distribution (No Mode):**
   - A uniform distribution has no clear mode.
   - The data is equally distributed across the entire range.
   - Example: Continuous random variable with a flat probability distribution.

Understanding the number of modes in a histogram provides insights into the underlying patterns and characteristics of the data. Unimodal distributions are common in many natural phenomena, while bimodal or multimodal distributions may indicate a more complex structure or the presence of multiple subpopulations. Uniform distributions, on the other hand, suggest randomness or equal likelihood across the variable's range.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Create a figure with four subplots in a single row
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Example 1: Unimodal Distribution (Normal distribution)
unimodal_data = np.random.normal(loc=0, scale=1, size=1000)
axes[0].hist(unimodal_data, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Unimodal Distribution (Normal)')

# Example 2: Bimodal Distribution (Mixture of two normal distributions)
bimodal_data = np.concatenate([np.random.normal(loc=-2, scale=1, size=500),
                                np.random.normal(loc=2, scale=1, size=500)])
axes[1].hist(bimodal_data, bins=30, color='salmon', edgecolor='black')
axes[1].set_title('Bimodal Distribution')

# Example 3: Multimodal Distribution (Mixture of several normal distributions)
multimodal_data = np.concatenate([np.random.normal(loc=-3, scale=1, size=300),
                                  np.random.normal(loc=0, scale=1, size=300),
                                  np.random.normal(loc=3, scale=1, size=400)])
axes[2].hist(multimodal_data, bins=30, color='lightgreen', edgecolor='black')
axes[2].set_title('Multimodal Distribution')

# Example 4: Uniform Distribution (No Mode)
uniform_data = np.random.uniform(low=-2, high=2, size=1000)
axes[3].hist(uniform_data, bins=30, color='lightcoral', edgecolor='black')
axes[3].set_title('Uniform Distribution')

# Set common labels for all subplots
for ax in axes:
    ax.set_xlabel('Values')
    ax.set_ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Generate example data with outliers
normal_data_with_outliers = np.concatenate([np.random.normal(loc=0, scale=1, size=950),
                                            np.random.uniform(low=-5, high=5, size=50)])

uniform_data_with_outliers = np.concatenate([np.random.uniform(low=-2, high=2, size=950),
                                              np.random.uniform(low=-10, high=10, size=50)])

skewed_data_with_outliers = np.concatenate([np.random.normal(loc=-2, scale=1, size=800),
                                            np.random.normal(loc=2, scale=1, size=800),
                                            np.random.uniform(low=-10, high=10, size=50)])

# Create a figure with subplots in a single column
fig, axes = plt.subplots(3, 1, figsize=(8, 15))

# Box plot for Normal Distribution with Outliers
boxplot1 = axes[0].boxplot(normal_data_with_outliers, labels=['Normal'], patch_artist=True)
axes[0].set_title('Distribution with Outliers')

# Box plot for Uniform Distribution with Outliers
boxplot2 = axes[1].boxplot(uniform_data_with_outliers, labels=['Uniform'], patch_artist=True)
axes[1].set_title('Distribution with Outliers')

# Box plot for Skewed Distribution with Outliers
boxplot3 = axes[2].boxplot(skewed_data_with_outliers, labels=['Skewed'], patch_artist=True)
axes[2].set_title('Distribution with Outliers')

# Set common labels for all subplots
for ax in axes:
    ax.set_ylabel('Values')

# Add legend and data labels
legend_labels = ['Normal', 'Uniform', 'Skewed']
for boxplot, label in zip([boxplot1, boxplot2, boxplot3], legend_labels):
    boxplot['boxes'][0].set_facecolor('lightblue')  # Customize box color
    boxplot['medians'][0].set_color('red')  # Customize median line color
    boxplot['fliers'][0].set(marker='+', color='green', alpha=0.5)  # Customize outlier marker
    boxplot['whiskers'][0].set_color('gray')  # Customize whisker color
    boxplot['whiskers'][1].set_color('gray')  # Customize whisker color
    boxplot['caps'][0].set_color('gray')  # Customize cap color
    boxplot['caps'][1].set_color('gray')  # Customize cap color

# Add legend
fig.legend([boxplot["boxes"][0] for boxplot in [boxplot1, boxplot2, boxplot3]],
           legend_labels, loc='upper left')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

# Scatter Plots

Scatter plots are graphical representations that help visualize the relationship between two continuous variables. They provide insights into patterns, trends, and correlations within the data. Key aspects of scatter plots include:

1. **Relationship between Two Continuous Variables:**
   - Scatter plots reveal the connection between two continuous variables, allowing for a visual examination of how changes in one variable correspond to changes in the other.

2. **Patterns, Trends, and Correlation:**
   - By examining the distribution of data points, patterns and trends in the relationship between variables become apparent. The visual representation aids in understanding the nature and strength of the correlation.

3. **Identification of Outliers:**

4. **Understanding Strength and Direction of Relationship:**

Using scatter plots, one can gain valuable insights into the nature of the data, identify potential outliers, and assess the strength and direction of the relationship between two continuous variables.

# Outliers

Outliers are data points that significantly deviate from the overall pattern of a dataset. Detecting and handling outliers is crucial for accurate analysis and model performance.

## How to Detect Outliers

1. **Visual Inspection:**
   - Scatter plots, box plots, and histograms can reveal the presence of outliers.
   - Visualize the distribution to identify data points that deviate from the main cluster.

2. **Statistical Methods:**
   - Z-Score: Identify data points beyond a certain number of standard deviations from the mean.
   - IQR (Interquartile Range): Identify data points outside the range defined by Q1 - 1.5 * IQR and Q3 + 1.5 * IQR.

3. **Machine Learning Models:**
   - Use clustering or anomaly detection algorithms to automatically identify outliers.

## How to Remove Outliers

1. **Z-Score Method:**
   - Exclude data points with z-scores beyond a chosen threshold (e.g., 3 standard deviations).

2. **IQR Method:**
   - Remove data points outside the range defined by Q1 - 1.5 * IQR and Q3 + 1.5 * IQR.

3. **Percentile Method:**
   - Exclude a certain percentage of data points from the lower and upper tails of the distribution.

4. **Machine Learning Models:**
   - Train models robust to outliers or use outlier-resistant algorithms.

## Considerations

1. **Impact on Analysis:**
   - Assess the impact of removing outliers on the overall analysis and interpretation of results.

2. **Domain Knowledge:**
   - Consider domain knowledge to determine whether outliers are valid data points or errors.

3. **Data Integrity:**
   - Ensure that removing outliers does not compromise the integrity of the dataset.

4. **Documentation:**
   - Document the process of outlier detection and removal for transparency.

Handling outliers appropriately is essential for obtaining reliable insights from data. Choose the method that aligns with the characteristics of your dataset and the goals of your analysis.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Generate example data
x = np.linspace(0, 10, 100)
y = 2 * x + np.random.normal(0, 2, 100)  # Linear relationship with some noise

# Introduce an outlier
y[10] = 30

# Create a figure with subplots in two rows
fig, axes = plt.subplots(2, 4, figsize=(18, 8))

# Scatter Plot (Original Data with Outlier)
axes[0, 0].scatter(x, y, label='Original Data Points (with Outlier)', color='red', alpha=0.7)
axes[0, 0].set_title('Scatter Plot (Original Data with Outlier)')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].legend()

# Box Plot (Original Data with Outlier)
data_for_boxplot = np.concatenate([np.random.normal(loc=0, scale=1, size=50),
                                   np.random.normal(loc=5, scale=1, size=50)])
axes[0, 1].boxplot(data_for_boxplot, labels=['Box Plot (Original Data with Outlier)'], patch_artist=True)
axes[0, 1].set_title('Box Plot (Original Data with Outlier)')
axes[0, 1].set_ylabel('Values')

# Left-skewed Histogram (Vertical, Original Data with Outlier)
left_skewed_data = np.random.exponential(scale=2, size=100)
axes[0, 2].hist(left_skewed_data, bins=20, color='green', alpha=0.7)
axes[0, 2].set_title('Left-Skewed Histogram (Original Data with Outlier)')
axes[0, 2].set_xlabel('Values')
axes[0, 2].set_ylabel('Frequency')

# Right-skewed Histogram (Vertical, Original Data with Outlier)
right_skewed_data = np.random.chisquare(df=3, size=100)
axes[0, 3].hist(right_skewed_data, bins=20, color='orange', alpha=0.7)
axes[0, 3].set_title('Right-Skewed Histogram (Original Data with Outlier)')
axes[0, 3].set_xlabel('Values')
axes[0, 3].set_ylabel('Frequency')

# Remove Outliers
outlier_index = np.argmax(y)
x_no_outlier = np.delete(x, outlier_index)
y_no_outlier = np.delete(y, outlier_index)

# Scatter Plot (Data without Outlier)
axes[1, 0].scatter(x_no_outlier, y_no_outlier, label='Data Points (without Outlier)', color='blue', alpha=0.7)
axes[1, 0].set_title('Scatter Plot (Data without Outlier)')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')
axes[1, 0].legend()

# Box Plot (Data without Outlier)
data_no_outlier = np.copy(data_for_boxplot)
data_no_outlier[outlier_index % 50] = 30  # Introduce an outlier in the same position
axes[1, 1].boxplot(data_no_outlier, labels=['Box Plot (Data without Outlier)'], patch_artist=True)
axes[1, 1].set_title('Box Plot (Data without Outlier)')
axes[1, 1].set_ylabel('Values')

# Left-skewed Histogram (Vertical, Data without Outlier)
left_skewed_data_no_outlier = np.random.exponential(scale=2, size=99)
axes[1, 2].hist(left_skewed_data_no_outlier, bins=20, color='green', alpha=0.7)
axes[1, 2].set_title('Left-Skewed Histogram (Data without Outlier)')
axes[1, 2].set_xlabel('Values')
axes[1, 2].set_ylabel('Frequency')

# Right-skewed Histogram (Vertical, Data without Outlier)
right_skewed_data_no_outlier = np.random.chisquare(df=3, size=99)
axes[1, 3].hist(right_skewed_data_no_outlier, bins=20, color='orange', alpha=0.7)
axes[1, 3].set_title('Right-Skewed Histogram (Data without Outlier)')
axes[1, 3].set_xlabel('Values')
axes[1, 3].set_ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

# Covariance and Correlation

## Covariance:

Covariance measures how two variables change together. It provides information about the direction of the relationship (positive or negative) but does not give a standardized measure of the strength of the relationship.

The formula for the covariance between two variables X and Y is given by:

\[ \text{Cov}(X, Y) = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{n-1} \]

Where:
- \(X_i\) and \(Y_i\) are individual data points for variables X and Y.
- \(\bar{X}\) and \(\bar{Y}\) are the means of variables X and Y, respectively.
- \(n\) is the number of data points.

## Correlation:

Correlation standardizes the measure of the relationship between two variables, providing a value between -1 and 1. The sign indicates the direction (positive or negative), and the magnitude indicates the strength of the relationship. A correlation of 1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, and 0 indicates no linear relationship.

The formula for the correlation coefficient (\(r\)) between X and Y is given by:

\[ r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y} \]

Where:
- \(\text{Cov}(X, Y)\) is the covariance between X and Y.
- \(\sigma_X\) and \(\sigma_Y\) are the standard deviations of X and Y, respectively.

### Key Points:

- Covariance can take any value, positive or negative, while correlation is always between -1 and 1.
- A correlation of 0 does not necessarily imply no relationship, as it only measures linear relationships.

In summary, covariance indicates the direction of the relationship, and correlation provides a standardized measure of both direction and strength. Correlation is often preferred in practice as it is easier to interpret and compare across different pairs of variables.


```python
import numpy as np

# Example Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])

# Covariance
covariance = np.cov(x, y, ddof=0)[0, 1]
print(f'Covariance between x and y: {covariance}')

# Correlation
correlation = np.corrcoef(x, y)[0, 1]
print(f'Correlation between x and y: {correlation}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])

# Scatter Plot
plt.scatter(x, y, label='Data Points', color='blue')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Correlation Coefficient
correlation = np.corrcoef(x, y)[0, 1]
print(f'Correlation between x and y: {correlation}')

# Optional: Regression Line
coefficients = np.polyfit(x, y, 1)
line = np.polyval(coefficients, x)

# Scatter Plot with Regression Line
plt.scatter(x, y, label='Data Points', color='blue')
plt.plot(x, line, label='Regression Line', color='red', linestyle='--')
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Causation

Causation refers to the relationship between cause and effect, where one event (the cause) brings about another event (the effect). It's important to note that establishing causation is a complex task and requires more than just observing a correlation between two variables.

## Key Points:

1. **Correlation does not imply causation:**
   - Just because two variables are correlated does not mean that one variable causes the other. Correlation only measures the strength and direction of a linear relationship between variables.

2. **Criteria for Establishing Causation:**
   - To establish causation, researchers often rely on experiments, controlled studies, and statistical methods to rule out alternative explanations. Some key criteria include:
      - **Temporal Order:** The cause must precede the effect in time.
      - **Correlation:** There should be a correlation between the cause and effect.
      - **Non-Spuriousness:** The observed relationship should not be due to a third variable.

3. **Experimental Studies:**
   - Randomized controlled trials (RCTs) are often considered the gold standard for establishing causation. In an experiment, researchers manipulate the independent variable (the potential cause) and observe its effect on the dependent variable (the potential effect).

4. **Confounding Variables:**
   - Confounding variables are external factors that may affect both the cause and the effect, leading to a spurious correlation. Controlling for confounding variables is crucial when trying to establish causation.

5. **Counterfactual Framework:**
   - Causation is often discussed within a counterfactual framework, where researchers compare what actually happened with what would have happened in the absence of the cause.

6. **Directionality Problem:**
   - In some cases, it is challenging to determine which variable is the cause and which is the effect. This is known as the directionality problem.

7. **Strength of Association:**
   - A strong association between two variables does not necessarily imply causation. Other factors, such as study design and the presence of confounding variables, need to be considered.

In summary, establishing causation requires careful consideration of various factors, and researchers often use a combination of study designs and statistical methods to draw meaningful conclusions about cause-and-effect relationships.
